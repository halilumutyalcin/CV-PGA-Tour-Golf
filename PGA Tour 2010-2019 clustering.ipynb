```python
#Finally, converting all values from string to numerical
PGAdf['Value'] = pd.to_numeric(PGAdf['Value'])
```

# Clustering

## KMeans clustering

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_predict = kmeans.fit_predict(X)
type(y_predict)
```

```python
# pd.merge(y, pd.DataFrame(y_predict))
y = pd.DataFrame(y)
y_predict = pd.DataFrame(y_predict)
merged = y.merge(y_predict, left_on=y.index, right_on=y_predict.index)
merged = merged.drop('key_0', 1)
PGAdfFinal = merged.merge(X, left_on=merged.index, right_on=X.index)
```

```python
PGAdfFinal = PGAdfFinal.rename(columns={0: 'Cluster'})
```

```python
cols = PGAdfFinal.columns[3:]
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
PGAdfFinal_reduced = pca.fit_transform(PGAdfFinal[cols])
PGAdfFinal['x'] = pca.fit_transform(PGAdfFinal[cols])[:, 0]
PGAdfFinal['y'] = pca.fit_transform(PGAdfFinal[cols])[:, 1]
PGAdfFinal
```
